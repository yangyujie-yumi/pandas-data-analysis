```python
df2=df[(df["Quantity"]>0)&(df["UnitPrice"]>0)]
```

```python
R_value=df2.groupby("CustomerID").date.max()
```

```python
R_value=df2["date"].max()-R_value
```

```python
R_value=R_value/np.timedelta64(1,"D")
```

```python
F_value=df2.groupby("CustomerID")["InvoiceNo"].nunique()
```

```python
M_value=df2.groupby("CustomerID")["amount"].sum()
```

```python
R_bins=[0,30,90,180,360,720]
```

```python
R_score=pd.cut(R_value,R_bins,labels=[5,4,3,2,1],right=False)
```

```python
F_bins=[1,2,5,10,20,5000]
```

```python
M_bins=[0,500,2000,5000,10000,200000]
```

```python
F_score=pd.cut(F_value,F_bins,labels=[1,2,3,4,5],right=False)
```

```python
M_score=pd.cut(M_value,M_bins,labels=[1,2,3,4,5],right=False)
```

```python
rfm=pd.concat([R_score,F_score,M_score],axis=1)
```

```python
rfm.columns=["R_score","F_score","M_score"]
```

```python
rfm.info()
```

```python
rfm["F_score"]=rfm["F_score"].astype(float)
```

```python
rfm["R_score"]=rfm["R_score"].astype(float)
```

```python
rfm["M_score"]=rfm["M_score"].astype(float)
```

```python
b=a.applymap(lambda x:"高" if x>0 else "低")
```

```python
def label(x):
    d={
        "低低高":"重要客户",
        "低低低":"重客户",
        "低高高":"重要户",
        "低高低":"重户",
        "高低高":"要客户",
        "高低低":"户",
        "高高高":"客",
        "高高低":"重要"
    }
    result=d[x]
    return result
    
```

```python
rfm["label"]=rfm["value"].apply(label)
```
