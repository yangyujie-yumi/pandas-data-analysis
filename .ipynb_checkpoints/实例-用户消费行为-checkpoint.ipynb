```python
rfm=df.pivot_table(index="user_id",
                  values=["order_dt","order_amount","order_products"],
                  aggfunc={"order_dt":"max",
                           "order_amount":"sum", 
                           "order_products":"sum"})
```

```python
rfm.head()
```

```python
a=rfm[["R","F","M"]].apply(lambda x:x-x.mean())
a
```

```python
rfm.rename(columns={"order_products":"F","order_amount":"M"},inplace=True)
```

```python
a=rfm[["R","F","M"]].apply(lambda x:x-x.mean())
a
```

```python
def rfm_func(x):
    level=x.apply(lambda x:"1" if x>=0 else "0")
    label=level.R+level.F+level.M
    d={"111":"hu",
       "110":"jih",
       "101":"hius",
       "100":"gdu",
       "001":"hdsf",
       "000":"ers",
       "010":"ersf",
       "011":"ytuvshg"   
    }
    result=d[label]
    return result
```

```python
rfm["label"]=a.apply(rfm_func,axis=1)
```

```python
pivoted_counts=df.pivot_table(index="user_id",columns="month",values="order_amount",aggfunc="count").fillna(0)
```

```python
pivoted_counts
```

```python
df_purchase=pivoted_counts.applymap(lambda x:1 if x>0 else 0)
```

```python
df_purchase
df_purchase.apply(lambda x:x-x.mean(),axis=1)
```

```python
def active_status(data):
    status=[]
    for i in range(18):
        if data[i]==0:
            if len(status)>0:
                if status[i-1]=="unreg":
                    status.append("unreg")
                else:
                    status.append("unactive")
            else:
                status.append("unreg")
        else:
            if len(status)==0:
                status.append("new")
            else:
                if status[i-1]=="unactive":
                    status.append("return")
                elif status[i-1]=="unreg":
                    status.append("new")
                else:
                    status.append("active")
    return status        
```

```python
purchase_stats=df_purchase.apply(active_status,axis=1)
```

```python
purchase_stats.reset_index()
```

```python
purchase_stats.columns=list(pivoted_counts.columns)
```
